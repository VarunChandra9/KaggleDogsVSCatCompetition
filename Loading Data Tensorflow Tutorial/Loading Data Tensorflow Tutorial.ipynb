```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import cv2
```

### Converting Data into grayscale

```python
# 1 for dog and 0 for cat
training_data = []
IMG_SIZE = 100
path = 'dogs-vs-cats\\train'
for img in os.listdir(path):
    try:
        img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        if img[:3] == 'cat':
            training_data.append([new_array, 0])
        else:
            training_data.append([new_array, 1])
    except Exception as e:
        pass
```

```python
print(len(training_data))
```

```python
training_data
```

```python
# We should shuffle the data because first half all of them are dogs and then we have cats
import random

random.shuffle(training_data)
```

```python
for sample in training_data:
    print(sample[1])
```

```python
X = []
y = []
```

```python
for features, label in training_data:
    X.append(features)
    y.append(label)
    
X = np.array(X).reshape(-1, IMG_SIZE, IMG_SIZE, 1) # 1 for GRAYSCALE, 3 for COLOR 
```

```python
import pickle

pickle_out = open("X.pickle", "wb")
pickle.dump(X, pickle_out)
pickle_out.close()

pickle_out = open("y.pickle", "wb")
pickle.dump(y, pickle_out)
pickle_out.close()
```
